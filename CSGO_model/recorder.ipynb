# Recorder
This records screen of resolution 1920x1080@60 for 100sec.

## Object classification

```python
!git clone https://github.com/ultralytics/yolov5.git
```

```python
import torch
```

```python
model = torch.hub.load('yolov5', 'yolov5s', source =  'local', force_reload = True)
```

```python
model
```

## Capturing Screen Data

```python
import pyautogui
import cv2
import numpy as np
import time
import os
```

```python
dir = './dataset'
os.makedirs(dir, exist_ok=True)
print("Saving images to:", os.path.abspath(dir))
```

```python
n = 1

while True:

    screen = pyautogui.screenshot()
    screen_array = np.array(screen)
    cropped_region = screen_array
    corrected_colors = cv2.cvtColor(cropped_region, cv2.COLOR_RGB2BGR)

    results = model(corrected_colors)

    filename = os.path.join(dir, f"{n:05d}.jpg")

    cv2.imwrite(filename, np.squeeze(corrected_colors))
    cv2.imshow('YOLO', np.squeeze(results.render()))

    n += 1

    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
# Close down the frame
cv2.destroyAllWindows()
```
